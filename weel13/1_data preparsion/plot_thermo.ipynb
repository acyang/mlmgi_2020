```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import pandas as pd
import math
```

```python
with open('tdo.txt','r') as csvfile:
    thermo_output=pd.read_csv(csvfile, delimiter=' ')
    
thermo_output.head()

print(thermo_output.columns)
print(thermo_output.keys())
```

```python
feature=thermo_output[['Temp', 'Press', 'Volume', 'PotEng', 'KinEng', 'TotEng', 'Enthalpy']]
for item in feature.columns:
    thermo_output[['Time',item]].plot(kind = 'line', x ='Time', y=item, title = 'Time vs. '+item, legend = False)
```
